```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.drop(['RowNumber', 'CustomerId','Surname'], axis = 1, inplace=True)
```

```python
df
```

```python
df_geo = pd.get_dummies(df['Geography'],dtype='int')
df_gender = pd.get_dummies(df['Gender'],dtype='int')
```

```python
df_one_hot = pd.concat([df_geo,df_gender,df], axis=1)
```

```python
df_one_hot.head()
```

```python
df_one_hot.drop(['Geography','Gender'], axis = 1, inplace=True)
```

```python
numerical_cols = [col for col in df_one_hot.columns if df_one_hot[col].dtype != 'object']

for col in numerical_cols:
    sns.kdeplot(df_one_hot[col])
    plt.title(f'KDE of {col}')
    plt.show()
```

```python
X = df_one_hot.iloc[:,:-1]
y = df_one_hot.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50, batch_size=10, verbose=1)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
